```python
import pandas as pd
df= pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Adex/Base_Ene_MC.xlsx', sheet_name='Base_Ene_MC')
df
```

```python
# Filtrar los datos por el año 2023 y ventas mayores o iguales a 1000
df_filtro = df[(df['AA'] == 2023) & (df['FOB'] >= 1000)]

# Agrupar por DECLARANTE, RUC_EMPRESA y Tamaño 2023, y contar los productos y destinos distintos
df_grouped = df_filtro.groupby(['NUM_DOCUMENTO', 'RUC_EMPRESA', 'Tamaño 2023']).agg(
    Num_Productos=('PARTIDA', 'nunique'),
    Num_Destinos=('NOMB_PAIS_ES','nunique')
    )
df_grouped
```

```python
# Ordenar por Tamaño 2023
df_sorted = df_grouped.sort_values(by='Tamaño 2023')
  #df_filtro_grupos.sort_values(by='Tamaño 2023')

# Filtrar por Num_Productos y Num_Destinos igual a 1
df_filtered_num_products = df_sorted[
    (df_sorted['Num_Productos'] == 1) & (df_sorted['Num_Destinos'] == 1)
    ]

# Imprimir los resultados
df_filtered_num_products
```
