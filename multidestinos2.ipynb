```python
#OPCION DE CARGA N° 2
#Matriz_MC es la matriz que contiene a las empresas ya clasificadas por Tamaño
import pandas as pd
##df= pd.read_excel('/content/drive/MyDrive/NOMBRE_CARPETA_DRIVE/NOMRE_EXCEL.xlsx', sheet_name='NOMBRE_HOJA_EXCEL')
df= pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Adex/Base_Ene_MC.xlsx', sheet_name='Base_Ene_MC')
df
```

```python
def multidestinos(df, anio: int=None, tipo=0):

    """
    Muestra el número de empresas exportadoras por país de destino.
    :df: DataFrame.
    :anio: Último año del dataframe (opcional, por defecto año actual).
    :tipo: Tipo de exportación (0: anual acumulado, 1: mes específico).
    """
    if tipo not in (0, 1):
      print("Error: el argumento 'tipo' solo acepta los valores 0 (acumulado anual) o 1 (mensual).")
      return

    #Si no se declara argumento 'anio', se utiliza, por defecto, el año actual.
    min_AA= min(df["AA"].unique())
    max_AA= max(df["AA"].unique())
    if anio is None:
      anio = max_AA
    elif not isinstance(anio, int) or anio<min_AA or anio>max_AA:
      print(f"Error: en este caso, a partir del análisis de la base de datos, el argumento 'anio' solo acepta valores enteros, desde {min_AA} hasta {max_AA}.")
      return

    #Determina cuántos años (máx. 10) se analizarán contando a partir del último año del dataframe.
    intentos2=3
    while True:
      intentos2-=1
      if intentos2==0:
        print("Excedió el límite de intentos.")
        return
      try:
        n = int(input(f"Ingrese el número de años a analizar, contando desde el argumento 'anio' (en este caso: {anio}): "))
        if 1<=n<=(anio - min_AA + 1):
          break # Sale del bucle si n es válido
        else:
          print(f'En este caso, a partir del análisis de la base de datos, el número de años a analizar debe ser menor o igual a {anio-min_AA+1}.')
      except ValueError:
        print('Error: el número de años a analizar debe ser un número entero.')

    #Creamos diccionario vacío para almacenar "SUB" DataFrames filtrados por año.
    df_dict = {}
    if tipo==0 :
      for i in range(0,n):
        df_dict[f'df{anio-i}'] = df[df['AA'] == anio-i]
    elif tipo==1 :
      intentos3=3
      min_MM=min(df["MM"].unique())
      max_MM=max(df["MM"].unique())
      while True:
        intentos3-=1
        if intentos3==0:
          print("Excedió el límite de intetos.")
          return
        try:
          mes = int(input(f"Ingrese el mes  (a partir del análisis de la base de datos: [{min_MM}, {max_MM}]): "))
          if min_MM <= mes <= max_MM:
              break  # Salir del bucle si el mes es válido
          else:
              print(f'En este caso, a partir del análisis de la base de datos, el mes debe ser un número entero, desde {min_MM} hasta {max_MM}.')
        except ValueError:
          print('Error: el mes debe ser un número entero.')

      for i in range(0,n):
          df_dict[f'df{anio-i}'] = df[(df['AA'] == anio-i) & (df['MM']==mes)]

    #Aplicación de filtros al DataFrame.
    import pandas as pd
    #Creamos diccinario vacío para almacenar Dataframes filtrados por año y FOB.
    multidestinos_dict = {}
    for i in range(0,n):
      #Agrupa por empresa, país y tamaño, y calcula la suma de ventas.
      f_multidestinos = df_dict[f'df{anio-i}'].groupby(
          ['NUM_DOCUMENTO', 'COD_PAIS', f'Tamaño {anio-i}'], as_index=False).agg({'FOB': 'sum'})
      #Calcula las ventas por empresa utilizando el DataFrame filtrado original.
      ventas_por_empresa = df_dict[f'df{anio-i}'].groupby('NUM_DOCUMENTO')['FOB'].sum().reset_index()
      ventas_por_empresa = ventas_por_empresa.rename(columns={'FOB': 'VentaTotal'})
      #Fusiona los DataFrames para combinar las ventas por empresa.
      f_multidestinos = pd.merge(f_multidestinos, ventas_por_empresa, on='NUM_DOCUMENTO')
      #Selecciona las columnas necesarias
      f_multidestinos = f_multidestinos[['NUM_DOCUMENTO', 'COD_PAIS', f'Tamaño {anio-i}', 'FOB', 'VentaTotal']]
      f_multidestinos.reset_index(drop=True, inplace=True)
      #Filtra los datos de empresas cuyas ventas totales sean >=1000 y, a su vez, que sus ventas por país >=1000
      #Nota: supongamos que una empresa exporta $1000: $500 Argentina y $500 Chile, entonces, esta empresa se considerará
      #en el conteo de empresas exportadoras (>=1000), PERO, NO se considerará en los destinos Argentina y Chile.
      f_multidestinos = f_multidestinos.loc[(f_multidestinos['VentaTotal'] >= 1000) & (f_multidestinos['FOB'] >= 1000)]

      #Inserta "SUB"-dataframes al diccionario.
      if tipo==0 :
        #Extrae columnas RUC, PAIS, Tamaño, y agrupa por PAIS y Tamaño. Luego, cuenta empresas existentes por PAIS y Tamaño.
        multidestinos_dict[f'{anio-i}'] = f_multidestinos.loc[
            :,['NUM_DOCUMENTO', 'COD_PAIS', f'Tamaño {anio-i}']
            ].groupby(
            ['COD_PAIS', f'Tamaño {anio-i}'],as_index=False
            )['NUM_DOCUMENTO'].count()
        #Cambio de nombre Tamaño {anio-i} a 'Tamanio'
        #Importante: este nombre debe coincidir con el DF utilizado en el Merge final
        multidestinos_dict[f'{anio-i}'].rename(
            columns={f'Tamaño {anio-i}':'Tamanio','NUM_DOCUMENTO':f'Empresas_{anio-i}'}, inplace=True)
      else :
        #Extrae columnas RUC y PAIS, y agrupa por PAIS. Luego, cuenta empresas existentes por PAIS.
        intentos4 = 4
        while True:
          intentos4-=1
          if intentos4==0:
            print("Excedió el límite de intentos.")
            return
          try:
            rpta = str(input(f"¿Desea incluir la clasificación por tamaño? (solo escriba: SI o NO): "))
            if rpta == 'SI':
              multidestinos_dict[f'{anio-i}'] = f_multidestinos.loc[
                  :, ['NUM_DOCUMENTO', 'COD_PAIS',f'Tamaño {anio-i}']
                  ].groupby(
                      ['COD_PAIS', f'Tamaño {anio-i}'], as_index=False
                      )['NUM_DOCUMENTO'].count()
              multidestinos_dict[f'{anio-i}'].rename(
                  columns={f'Tamaño {anio-i}':'Tamanio','NUM_DOCUMENTO':f'Empresas_{anio-i}_{mes}'}, inplace=True
                  )
              break  # Salir del bucle si el mes es válido
            elif rpta =='NO':
              multidestinos_dict[f'{anio-i}'] = f_multidestinos.loc[
                  :, ['NUM_DOCUMENTO', 'COD_PAIS']
                  ].groupby(
                      ['COD_PAIS'], as_index=False
                      )['NUM_DOCUMENTO'].count()
              multidestinos_dict[f'{anio-i}'].rename(
                  columns={'NUM_DOCUMENTO':f'Empresas_{anio-i}_{mes}'}, inplace=True
                  )
              break
          except ValueError:
            print('Error: solo debe escribir "SI" o "NO"')

    #Construyendo el Merge Final.
    from itertools import product
    #Se crea la tupla 'paises' (no modificable post creación)
    paises= (
        '1A','1B','1C','1D','1E','1F','1G','1H','1I','1J','1K','1L','1M','1N','1O','1P','1Q','1R','1S','1T','1U','1V','1W','99',
        'AD','AE','AF','AG','AI','AL','AM','AN','AO','AQ','AR','AS','AT','AU','AW','AX','AZ','BA','BB','BD','BE','BF','BG','BH',
        'BI','BJ','BL','BM','BN','BO','BQ','BR','BS','BT','BV','BW','BY','BZ','CA','CC','CD','CF','CG','CH','CI','CK','CL','CM',
        'CN','CO','CR','CS','CU','CV','CW','CX','CY','CZ','DE','DJ','DK','DM','DO','DZ','EC','EE','EG','EH','ER','ES','ET','FI',
        'FJ','FK','FM','FO','FR','FX','GA','GB','GD','GE','GF','GG','GH','GI','GL','GM','GN','GP','GQ','GR','GS','GT','GU','GW',
        'GY','HK','HM','HN','HR','HT','HU','ID','IE','IL','IM','IN','IO','IQ','IR','IS','IT','JE','JM','JO','JP','KE','KG','KH',
        'KI','KM','KN','KP','KR','KW','KY','KZ','LA','LB','LC','LI','LK','LO','LR','LS','LT','LU','LV','LY','MA','MC','MD','ME',
        'MF','MG','MH','MK','ML','MM','MN','MO','MP','MQ','MR','MS','MT','MU','MV','MW','MX','MY','MZ','NA','NC','NE','NF','NG',
        'NI','NL','NM','NO','NP','NR','NT','NU','NZ','OM','PA','PE','PF','PG','PH','PK','PL','PM','PN','PR','PS','PT','PW','PY',
        'QA','RE','RO','RS','RU','RW','SA','SB','SC','SD','SE','SG','SH','SI','SJ','SK','SL','SM','SN','SO','SR','SS','ST','SV',
        'SX','SY','SZ','TC','TD','TF','TG','TH','TJ','TK','TL','TM','TN','TO','TP','TR','TT','TV','TW','TZ','UA','UG','UM','US',
        'UY','UZ','VA','VC','VE','VG','VI','VN','VU','WF','WS','XZ','YE','YT','YU','ZA','ZM','ZR','ZW'
        )
    #Se crea la tupla 'tamanos' (no modificable post creación)
    tamanos= (
        'Grande',
        'Mediana',
        'Pequeña',
        'Microempresa'
        )

    if tipo==0 :
      #Se crea el producto cartesiano entre paises y tamanos.
      #Importante: la columna 'Tamanio' debe coincidir con el nombre del if tipo==0 del 1° for
      pxt=pd.DataFrame(product(paises,tamanos),columns=['COD_PAIS','Tamanio'])
      #Merge entre pxt y los "SUB" dataframes de multidestinos_dict
      multidestinos=pd.merge(pxt, multidestinos_dict[f'{anio}'], how= 'outer', on=['COD_PAIS','Tamanio'])
      for j in range (1,n):
        multidestinos = pd.merge(
            multidestinos,
            multidestinos_dict[f'{anio-j}'],
            how= 'outer',
            on=['COD_PAIS','Tamanio']
            )
      #Se quitan las filas vacías. Se considera desde la columna 2, pues siempre estarán llenas ("COD_PAIS" y "Tamanio")
      multidestinos.dropna(subset=multidestinos.columns[2:], how="all", inplace = True)

    else :
      if rpta == 'SI':
        pxt=pd.DataFrame(product(paises,tamanos),columns=['COD_PAIS','Tamanio'])
        multidestinos=pd.merge(pxt, multidestinos_dict[f'{anio}'], how= 'outer', on=['COD_PAIS','Tamanio'])
        for j in range (1,n):
          multidestinos = pd.merge(
            multidestinos,
            multidestinos_dict[f'{anio-j}'],
            how= 'outer',
            on=['COD_PAIS','Tamanio']
            )
        multidestinos.dropna(subset=multidestinos.columns[2:], how="all", inplace = True)
      elif rpta == 'NO':
        pxt=pd.DataFrame(paises,columns=['COD_PAIS'])
        multidestinos=pd.merge(pxt, multidestinos_dict[f'{anio}'], how= 'outer', on=['COD_PAIS'])
        for j in range (1,n):
          multidestinos = pd.merge(
              multidestinos,
              multidestinos_dict[f'{anio-j}'],
              how= 'outer',
              on=['COD_PAIS']
              )
      #Se quitan las filas vacías. Se considera desde la columna 1, pues siempre estará llena ("COD_PAIS")
      multidestinos.dropna(subset=multidestinos.columns[1:], how="all", inplace = True )

    #Exporta a Excel.
    multidestinos.to_excel(f'multidestinos_tipo{tipo}.xlsx', index=False)
```
